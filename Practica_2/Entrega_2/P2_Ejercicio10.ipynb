# Ejercicio 10
Simule una competencia de cocina donde 5 participantes son evaluados por 3 jueces en cada ronda. Cada juez otorga un puntaje del 1 al 10. El puntaje de la ronda es la suma de los tres jueces.
Realizar las siguientes operaciones:
- Imprimir la tabla de posiciones luego de cada ronda para ver la progresión de la competencia.
- Cada ronda debe indicar quién fue el ganador (mayor puntaje en la ronda).
- Contabilizar la cantidad de rondas ganadas por cada participante.
- Imprimir la tabla final con: puntaje total acumulado, cantidad de rondas ganadas, mejor puntaje en una ronda y promedio por ronda. La tabla debe estar en orden decreciente por puntaje total.

### Inicializo los acumuladores

```python
total = {}
rounds_won = {}
best_round = {}
participants = list(rounds[0]['scores'].keys())

for p in participants:
    total[p] = 0
    rounds_won[p] = 0
    best_round[p] = 0
```

### Procesmos las rondas

```python
round_number = 1
for round in rounds:
    theme = round['theme']
    round_scores = {}

    for p in participants:
        score = get_score(round['scores'][p])
        round_scores[p] = score
        total[p] += score
        if score > best_round[p]:
            best_round[p] = score
    # Ganador de la ronda        
    winner = participants[0]
    winner_score = round_scores[participants[0]]
    for p in participants:
        if round_scores[p] > winner_score:
            winner = p
            winner_score = round_scores[p]
    rounds_won[winner] += 1

                        # Ordeno de mayor a menor
    sorted_round = sorted(round_scores.items(), key=lambda x: x[1], reverse=True)
    # Las tuplas para ordenar son (nombre, puntaje) entonces ordenamos en x[1]

                        # Imprimo la ronda
    print(f"Ronda {round_number} - {theme}:")
    print(f"Ganador: {winner} ({winner_score} pts)")
    print(f"{'Cocinero':}    {'Puntaje ronda':}     {'Acumulado'}")
    print("-" * 40)
    for p, score in sorted_round:
        print(f"{p:}     {score:}            {total[p]}")
    print()
    round_number += 1
```

### Ordeno e imprimo la tabla fianl

```python
final = sorted(total.items(), key=lambda x: x[1], reverse=True)

print("Tabla de posiciones final:")
print(f"{'Cocinero':}    {'Puntaje':}    {'Rondas ganadas':}    {'Mejor ronda':}    {'Promedio'}")
print("-" * 70)
for p, score in final:
    average = total[p] / len(rounds)
    print(f"{p:}         {score:}            {rounds_won[p]:}        {best_round[p]:}         {average}")
print("-" * 70)
```
