# Ejercicio 7
Solicite al usuario una lista de nombres de participantes (separados por coma). El programa debe asignar aleatoriamente a cada persona un amigo invisible, cumpliendo que nadie se tenga a sí mismo.
Validaciones:
- Debe haber al menos 3 participantes.
- No debe haber nombres duplicados (comparar ignorando mayúsculas y espacios extra).

```python
import random

names = input("Ingrese los participantes (separados por coma): ").split(",")
for i in range(len(names)):
    names[i] = names[i].strip()

names_lower = set()
has_duplicates = False
for name in names:
    if name.lower() in names_lower:
        has_duplicates = True
        break
    names_lower.add(name.lower())

if len(names) < 3:
    print("Error: debe haber al menos 3 participantes.")
elif has_duplicates:
    print("Error: hay nombres duplicados.")
else:
    assigned = names.copy()
    while True:
        random.shuffle(assigned)
        valid = True
        for i in range(len(names)):
            if names[i] == assigned[i]:
                valid = False
                break
        if valid:
            break

    print("Sorteo de amigo invisible:")
    for i in range(len(names)):
        print(f"{names[i]} -> {assigned[i]}")
```
