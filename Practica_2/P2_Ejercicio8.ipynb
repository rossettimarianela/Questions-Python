# Ejercicio 8
Implemente el cifrado César, que consiste en desplazar cada letra del alfabeto una cantidad fija de posiciones. El programa debe:
- Solicitar un mensaje al usuario.
- Solicitar un valor de desplazamiento (número entero).
- Mostrar el mensaje cifrado.
- Mostrar el mensaje descifrado a partir del cifrado (para verificar que funciona).
- Las letras deben rotar (después de la Z viene la A). Los caracteres que no sean letras (espacios, signos de puntuación, números) se mantienen sin cambios. Debe funcionar tanto con mayúsculas como con minúsculas, preservando el formato original.

```python
message = input("Ingrese un mensaje: ")
shift = int(input("Ingrese el desplazamiento: "))

alphabet_lower = "abcdefghijklmnñopqrstuvwxyz"
alphabet_upper = "ABCDEFGHIJKLMNÑOPQRSTUVWXYZ"
```

```python
encrypted = ""
for char in message:
    if char in alphabet_lower:
        index = alphabet_lower.index(char)
        encrypted += alphabet_lower[(index + shift) % 27] 
        # En la posicion index (donde esta la letra) pero si por ej es z, que sea 0 la posicion (27/27 resto = 0)
    elif char in alphabet_upper:
        index = alphabet_upper.index(char)
        encrypted += alphabet_upper[(index + shift) % 27]
    else:
        encrypted += char

# deciframos a partir del mensaje cifrado:
decrypted = ""
for char in encrypted:
    if char in alphabet_lower:
        index = alphabet_lower.index(char)
        decrypted += alphabet_lower[(index - shift) % 27] 
    elif char in alphabet_upper:
        index = alphabet_upper.index(char)
        decrypted += alphabet_upper[(index - shift) % 27]
    else:
        decrypted += char
print(f'Mensaje Cifrado: {encrypted}')
print(f'Mensaje Decifrado: {decrypted}')
```
