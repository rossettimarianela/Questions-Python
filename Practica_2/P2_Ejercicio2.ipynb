```python
text = """Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
Flat is better than nested.
Sparse is better than dense.
Readability counts.
Special cases aren't special enough to break the rules.
Although practicality beats purity.
Errors should never pass silently.
Unless explicitly silenced.
In the face of ambiguity, refuse the temptation to guess.
There should be one-- and preferably only one --obvious way
to do it.
Although that way may not be obvious at first unless you're
Dutch.
Now is better than never.
Although never is often better than *right* now.
If the implementation is hard to explain, it's a bad idea.
If the implementation is easy to explain, it may be a good
idea.
Namespaces are one honking great idea -- let's do more of
those!"""
```

### Contamos el total de líneas: 

```python
lines = text.split("\n")
total_lines = len(lines)
print(f'El total de líneas es: {total_lines}')
```

### Contamos el total de palabras: 

```python
words = text.split()
total_words = len(words)
print(f'El total de palabras es de: {total_words}')
```

### Sacamos el promedio de palabras por línea: 

```python
average = total_words / total_lines
print(f'El promedio de palabras por linea es: {average}')
```

### Ahora sacamos todas las líneas cuya cantidad de palabras esté por encima del promedio:

```python
# words_per_line: por cada renglón en la variable lines, agarra la linea, hace una lista split()
# por espacios y cuenta cuantos elementos tiene esa list
words_per_line = [len(line.split()) for line in lines]
above_average = []
for i in range(len(lines)):
    if words_per_line[i] > average:
        above_average.append(lines[i])
print(f'Lineas por encima del promedio {average} :')

# recorro el above_average e imprimo el item (el string) y la cantidad de palabras que tiene
for i in range(len(above_average)):
    print(f"- \"{above_average[i]}\" ({len(above_average[i].split())} Palabras)")
```

# Ejercicio 2
Dada la siguiente playlist con canciones y sus duraciones en formato "m:ss"
- Calcule la duración total de la playlist en formato Xm Ys, y encuentre la canción más larga y la más corta.
Nota: Deberá convertir las duraciones de string a valores numéricos para poder operar con ellas.

```python
playlist = [
{"title": "Bohemian Rhapsody", "duration": "5:55"},
{"title": "Hotel California", "duration": "6:30"},
{"title": "Stairway to Heaven", "duration": "8:02"},
{"title": "Imagine", "duration": "3:07"},
{"title": "Smells Like Teen Spirit", "duration": "5:01"},
{"title": "Billie Jean", "duration": "4:54"},
{"title": "Hey Jude", "duration": "7:11"},
{"title": "Like a Rolling Stone", "duration": "6:13"},
]
```

### Convierto las duraciones a String

```python
for song in playlist:
    # con esto accedemos a la seccion duration del diccionario la cual es ej: 5:55 y separamos en la lista el [5,55]
    parts = song["duration"].split(":")
    song["seconds"] = int(parts[0]) * 60 + int(parts[1]) # convierto el 5 en segundos, (paso todo a segundos) y añado al diccionario
```

### Duracion total, canción más larga y canción más corta

```python
total_duration = 0
shortest_song = playlist[0]
longest_song = playlist[0]

for song in playlist:
    total_duration += song["seconds"]
    if song["seconds"] > longest_song["seconds"]:
        longest_song = song
    if song["seconds"] < shortest_song["seconds"]:
        shortest_song = song
        
minutes = total_duration // 60
seconds = total_duration % 60

print(f'Duración total: {minutes}m {seconds}s ')
print(f'Canción más larga: \"{longest_song["title"]}\" ({longest_song["duration"]})')
print(f'Canción más corta: \"{shortest_song["title"]}\" ({shortest_song["duration"]})')
```
